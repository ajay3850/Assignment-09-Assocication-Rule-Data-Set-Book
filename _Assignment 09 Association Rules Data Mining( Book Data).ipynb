# Apriori Algorithm
1. Association rules with 10% Support and 70% confidence

```python
# With 10% Support
frequent_itemsets=apriori(bk,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
# with 70% confidence
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
## A leverage value of 0 indicates independence. Range will be [-1 1]
## A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]
```

```python
rules.sort_values('lift',ascending=False)
```

A lift ratio larger than 1.0 implies that the relationship between the antecedent and
the consequent is more significant than would be expected if the two sets were independent.
The larger the lift ratio, the more significant the association.

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules[rules.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```
